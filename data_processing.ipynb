### Yahoo Historical data (2016-2023) and Earning Calendars

```python
## Params
start_date = '2016-01-01'
end_date = '2023-08-31'
```

```python
tickers =['MSFT', 'AAPL', 'MDB']
```

```python
import yfinance as yf

open_result = pd.DataFrame()
close_result = pd.DataFrame()
high_result = pd.DataFrame()
low_result = pd.DataFrame()
volume_result = pd.DataFrame()
earnings_calendars = {}

for ticker in tickers:
    ticker_data = yf.Ticker(ticker)

    #historical data
    historical_data = ticker_data.history(start=start_date, end=end_date)
    open = historical_data['Open']
    close = historical_data['Close']
    high = historical_data['High']
    low = historical_data['Low']
    volume = historical_data['Volume']
    open_result[ticker] = open
    close_result[ticker] = close
    high_result[ticker] = high
    low_result[ticker] = low
    volume_result[ticker] = volume 
```
