```python
import yfinance as yf

earnigns_date_by_ticker = {}
for i in range(len(tickers)):
    print(i, len(tickers))
    ticker = tickers[i]
    ticker_data = yf.Ticker(ticker)
    earnings_data = ticker_data.get_earnings_dates(limit=60)
    if earnings_data is not None:
        earnigns_date_by_ticker[ticker] = earnings_data.index

    
```

```python
for ticker in earnigns_date_by_ticker:
    print(ticker)
```

```python
earnings_file_path = "./data/earnings_data.txt"
with open(earnings_file_path, "w") as fw:
    for ticker in earnigns_date_by_ticker:
        fw.write("%s:" % ticker)
        earnings_dates = earnigns_date_by_ticker[ticker].tolist()
        for date in earnings_dates:
            fw.write("%s," % date.strftime("%Y-%m-%d"))
        fw.write("\n")
    
```
